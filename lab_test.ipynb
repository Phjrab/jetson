```python
import cv2
import matplotlib.pyplot as plt

# 1. 카메라 열기
cap = cv2.VideoCapture(0)

# 2. 화면 한 장 읽기
ret, frame = cap.read()

if ret:
    # OpenCV는 BGR 순서로 색상을 읽으므로, 출력을 위해 RGB로 바꿉니다.
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    # 쥬피터 노트북 안에 이미지 표시
    plt.figure(figsize=(10, 6))
    plt.imshow(frame_rgb)
    plt.axis('off') # 격자 숨기기
    plt.show()
else:
    print("화면을 캡처할 수 없습니다.")

# 3. 카메라 자원 해제 (중요!)
cap.release()
```
